1. 画一张深证综合指数的价格折线图。（数据可从深交所网站获取）


```python
# 将深圳综合指数的收盘价格保存在list中 (399001.SZ)

list_tradedate = [
    20180201,
    20180202,
    20180205,
    20180206,
    20180207,
    20180208,
    20180209,
    20180212,
    20180213,
    20180214,
    20180222,
    20180223,
    20180226,
    20180227,
    20180228
]

list_szzhzs = [
    10864.342,
    10925.161,
    10836.252,
    10377.610,
    10246.974,
    10372.588,
    10001.227,
    10291.882,
    10362.432,
    10431.914,
    10658.937,
    10662.792,
    10895.562,
    10807.535,
    10828.730
]
```

### 转换成日期格式

```python
# convert int to datetime
import datetime

list_tradedate2 = []

for trade_date in list_tradedate:
    dt_tradedate = datetime.datetime.strptime(str(trade_date), '%Y%m%d')
    list_tradedate2.append(dt_tradedate)
    
```

### 画图

```python
# 导入python画图库matplotlib
import matplotlib.pyplot as plt  

# 指定图形大小
plt.figure(figsize=(12, 8))

# 画折线图
plt.plot(list_tradedate2, list_szzhzs)

# 标识标题及坐标轴信息
plt.title('shenzhen stock exchange index')
plt.xlabel('trade_date')
plt.ylabel('close price')

# 显示画图结果
plt.show()
```

```python
# 导入python画图库matplotlib
import matplotlib.pyplot as plt  

# 指定图形大小
plt.figure(figsize=(12, 8))

# 画柱状图
plt.bar(list_tradedate2, list_szzhzs)

# 标识标题及坐标轴信息
plt.title('shenzhen stock exchange index')
plt.xlabel('trade_date')
plt.ylabel('close price')

# 显示画图结果
plt.show()
```

有点问题，这个中间的日期是有间隔的，怎么能够去掉呢？

基本原理是：X轴的坐标如果想要连续的话，中间的间隔必须是1 （整数差1，或者日期差1天）

```python
# 导入python画图库matplotlib
import matplotlib.pyplot as plt  
import matplotlib.ticker as ticker

# 定义x轴
date_num = len(list_tradedate2)
axis_x = [n for n in range(date_num)]

# 定义x轴的格式函数
def format_date(x, pos=None):
    #保证下标不越界,很重要,越界会导致最终plot坐标轴label无显示
    thisind = int(x+0.5)
    if thisind < 0:
        thisind = 0
    elif thisind > date_num-1:
        thisind = date_num-1

    return list_tradedate2[thisind].strftime('%Y%m%d')

# 指定图形大小
fig = plt.figure(figsize=(12, 8))

# 画柱状图
ax = fig.add_subplot(1,1,1)
ax.bar(axis_x, list_szzhzs)

# 设置坐标轴显示
ax.xaxis.set_major_formatter(ticker.FuncFormatter(format_date))

# 标识标题及坐标轴信息
ax.set_title('shenzhen stock exchange index')
ax.set_xlabel('trade_date')
ax.set_ylabel('close price')

# 显示画图结果
plt.show()
```
