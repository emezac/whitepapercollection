## 糟糕，发生了什么问题？

这个图形明显不是我们。

问题发生在了X轴上，list_tradedate轴上送入的是整数20180102，需要送入的是一个日期

解决方案：将X轴的数据转换为日期类型

需要使用datetime库中的datetime对象的strptime方法

```python
# convert int to datetime
import datetime

list_tradedate2 = []

for trade_date in list_tradedate:
    dt_tradedate = datetime.datetime.strptime(str(trade_date), '%Y%m%d')
    list_tradedate2.append(dt_tradedate)
    
```

```python
# 导入python画图库matplotlib
import matplotlib.pyplot as plt  

# 指定图形大小
plt.figure(figsize=(12, 8))

# 画柱状图
plt.plot(list_tradedate2, list_szzs)

# 标识标题及坐标轴信息
plt.title('shanghai stock exchange index')
plt.xlabel('trade_date')
plt.ylabel('close price')

# 显示画图结果
plt.show()
```

## 画一张2017年中国GDP构成的饼图

```python
# 2017年GDP构成数据准备
# 数据来源：新闻报道 
# https://baijiahao.baidu.com/s?id=1593629333261604794&wfr=spider&for=pc

gdp_2017 = {
    "primary industry"   : 65468,
    "secondary industry" : 334623,
    "tertiary industry"  : 427032
}

# 分别取出 gdp类别列表 和 gdp值列表
labels = gdp_2017.keys()
values = gdp_2017.values()
```

```python
for k, v in gdp_2017.items():
    print(k, v)
```

```python
# 导入python画图库matplotlib
import matplotlib.pyplot as plt

# 设置图形大小
plt.figure(figsize=(6,6))

# 画饼图 
# startangle表示饼图的起始角度
plt.pie(values, 
        labels=labels, 
        autopct='%.1f%%', 
        startangle=90
    ) 

# 设置样式
plt.axis('equal')

# 显示图例
plt.legend()

# 显示图片
plt.show()
```

## Set - 没有值的dict
