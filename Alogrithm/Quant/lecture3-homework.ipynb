```python
# 将数据保存到list数据结构中
list_tradedate = [
    20180201,
    20180202,
    20180205,
    20180206,
    20180207,
    20180208,
    20180209,
    20180212,
    20180213,
    20180214,
    20180222,
    20180223,
    20180226,
    20180227,
    20180228
]

list_close  = [
    34.45,
    34.50,
    34.89,
    34.62,
    32.39,
    31.28,
    30.71,
    30.05,
    30.91,
    31.37,
    31.50,
    31.78,
    31.51,
    30.87,
    30.30
]

list_change  = [
    0.015027  ,
    0.001451  ,
    0.011304  ,
    -0.007739 ,
    -0.064414 ,
    -0.034270 ,
    -0.018223 ,
    -0.021491 ,
    0.028619  ,
    0.014882  ,
    0.004144  ,
    0.008889  ,
    -0.008496 ,
    -0.020311 ,
    -0.018465 ,
]
```

```python
# convert int to datetime
import datetime

list_tradedate2 = []

for trade_date in list_tradedate:
    dt_tradedate = datetime.datetime.strptime(str(trade_date), '%Y%m%d')
    list_tradedate2.append(dt_tradedate)
```

```python
# 导入python画图库matplotlib
import matplotlib.pyplot as plt  
import matplotlib.ticker as ticker

# 设置图片大小
fig = plt.figure(figsize=(8,6))

# 定义x轴
date_num = len(list_tradedate2)
axis_x = [n for n in range(date_num)]

# 定义x轴的格式函数
def format_date(x, pos=None):
    #保证下标不越界,很重要,越界会导致最终plot坐标轴label无显示
    thisind = int(x+0.5)
    if thisind < 0:
        thisind = 0
    elif thisind > date_num-1:
        thisind = date_num-1

    return list_tradedate2[thisind].strftime('%Y%m%d')


# 画柱状图
ax1 = fig.add_subplot(1,1,1)
ax1.bar(axis_x, list_close, label='close')
ax1.legend(loc='upper left')
ax1.set_xlabel('trade_date')
ax1.set_ylabel('close')

# 设置坐标轴显示
ax1.xaxis.set_major_formatter(ticker.FuncFormatter(format_date))

ax2 = ax1.twinx()
ax2.plot(axis_x, list_change, color='red', label='change')
ax2.legend(loc='upper right')
ax2.set_ylabel('change')
ax2.set_ylim(-0.1, 0.1)

# 设置坐标轴显示
ax2.xaxis.set_major_formatter(ticker.FuncFormatter(format_date))

# 标识标题及坐标轴信息
plt.title('600036.SH close / change')

# 显示画图结果
plt.show()
```
